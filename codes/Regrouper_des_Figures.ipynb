```python
from PIL import Image, ImageDraw, ImageFont

# Charger les images principales
image1 = Image.open("/home/wester/Desktop/SST/ERA5/images/csig_ISO_olr_Anom_rain_hebdomadaire_colonnes.png")
image2 = Image.open("/home/wester/Desktop/SST/ERA5/images/csig_ISO_u850_Anom_rain_hebdomadaire_colonnes.png")
image3 = Image.open("/home/wester/Desktop/SST/ERA5/images/csig_ISO_u200_Anom_rain_hebdomadaire_colonnes.png")

# Charger l'image de la barre de couleur (cbar)
cbar = Image.open("/home/wester/Desktop/MonDossier/Figures/cbar.png")

# Récupérer les dimensions des images (on suppose qu'elles ont toutes la même taille)
largeur, hauteur = image1.size
hauteur_titre = 50  # Hauteur pour les sous-titres
hauteur_cbar = cbar.size[1]  # Hauteur de la barre de couleur

# Créer une nouvelle image avec de l'espace pour les titres et la barre de couleur
nouvelle_image = Image.new(
    'RGB', 
    (largeur * 3, hauteur + hauteur_titre + hauteur_cbar), 
    color=(255, 255, 255)  # Fond blanc
)

# Dessiner les sous-titres
draw = ImageDraw.Draw(nouvelle_image)

# Utiliser une police par défaut ou une spécifique si disponible
try:
    font = ImageFont.truetype("arial.ttf", 35)  # Spécifier la police et la taille
except IOError:
    font = ImageFont.load_default()  # Si Arial n'est pas disponible, utiliser la police par défaut

# Ajouter les sous-titres
draw.text((largeur * 0.5 - 20, 30), "OLR", fill="black", font=font)  # Sous-titre colonne 1
draw.text((largeur * 1.5 - 20, 30), "U850", fill="black", font=font)  # Sous-titre colonne 2
draw.text((largeur * 2.5 - 20, 30), "U200", fill="black", font=font)  # Sous-titre colonne 3

# Coller les trois images dans la nouvelle image sous les sous-titres
nouvelle_image.paste(image1, (0, hauteur_titre))  # Image 1 dans la première colonne
nouvelle_image.paste(image2, (largeur, hauteur_titre))  # Image 2 dans la deuxième colonne
nouvelle_image.paste(image3, (largeur * 2, hauteur_titre))  # Image 3 dans la troisième colonne

# Calculer la position centrée pour la barre de couleur
position_cbar_x = (nouvelle_image.size[0] - cbar.size[0]) // 2  # Centrer horizontalement
position_cbar_y = hauteur + hauteur_titre  # Position verticale sous les images principales
nouvelle_image.paste(cbar, (position_cbar_x, position_cbar_y))  # Coller la barre de couleur centrée

# Sauvegarder et afficher l'image combinée
nouvelle_image.save("/home/wester/Desktop/MonDossier/Figures/correlation_coefficient_Wind_with_centered_cbar.png", "PNG")
nouvelle_image.show()
```
