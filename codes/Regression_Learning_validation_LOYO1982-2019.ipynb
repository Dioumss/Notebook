```python
# Fonction pour calculer le coefficient de corrélation avec gestion des NaN
def correlation_coefficient(y_true, y_pred):
    if np.var(y_true) == 0 or np.var(y_pred) == 0:
        return 0  # Si la variance est nulle, la corrélation est 0
    else:
        return pearsonr(y_true, y_pred)[0]
```

**1. Fonction pour préparer les données :**

```python
from sklearn.model_selection import LeaveOneGroupOut

def prepare_data(delai, rain_Index, years_in_data):
    """
    Prépare les données pour l'entraînement et le test en utilisant la validation Leave-One-Group-Out (LOYO).
    """
    logo = LeaveOneGroupOut()
    for train_idx, test_idx in logo.split(delai, groups=years_in_data):
        X_train, X_test = delai[train_idx].values, delai[test_idx].values
        y_train, y_test = rain_Index[train_idx].values, rain_Index[test_idx].values
        
        # Convertir en 2D si nécessaire
        X_train = X_train.reshape(-1, 1) if X_train.ndim == 1 else X_train
        X_test = X_test.reshape(-1, 1) if X_test.ndim == 1 else X_test
        
        yield X_train, X_test, y_train, y_test
```

**2. Fonction pour entraîner le modèle et évaluer les performances :**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

def train_and_evaluate(X_train, X_test, y_train, y_test):
    """
    Entraîne un modèle de régression linéaire et évalue les performances.
    """
    # Création d'un objet Linear Regression
    model = LinearRegression()
    
    # Entraînement du modèle
    model.fit(X_train, y_train)
    
    # Prédictions
    y_pred = model.predict(X_test)
    
    # Calcul des scores
    mae = mean_absolute_error(y_test, y_pred)
    corr = np.corrcoef(y_test, y_pred)[0, 1]
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    return y_pred, mae, corr, rmse
```

**3. Fonction principale pour exécuter la validation et collecter les résultats :**

```python
def main(delais, delai_names, rain_Index):
    """
    Exécute le processus de validation LOYO pour plusieurs délais et collecte les résultats.
    """
    years_in_data = rain_Index.coords['time'].dt.year.values
    
    # Initialisation des listes pour stocker les résultats
    model_scores_mae, model_scores_corr, model_scores_rmse, predictions_by_delay = [], [], [], []
    
    # Boucle pour chaque délai
    for delai, name in zip(delais, delai_names):
        year_scores_mae, year_scores_corr, year_scores_rmse, predictions_for_delay = [], [], [], []
        
        # Préparation des données et boucle LOYO pour chaque année
        for X_train, X_test, y_train, y_test in prepare_data(delai, rain_Index, years_in_data):
            # Entraîner et évaluer le modèle
            y_pred, mae, corr, rmse = train_and_evaluate(X_train, X_test, y_train, y_test)
            
            predictions_for_delay.append(y_pred)
            year_scores_mae.append(mae)
            year_scores_corr.append(corr)
            year_scores_rmse.append(rmse)
        
        # Stockage des résultats pour ce délai
        model_scores_mae.append(year_scores_mae)
        model_scores_corr.append(year_scores_corr)
        model_scores_rmse.append(year_scores_rmse)
        predictions_by_delay.append(predictions_for_delay)
    
    return model_scores_mae, model_scores_corr, model_scores_rmse, predictions_by_delay
```

```python
# Appel de la fonction principale
model_scores_mae, model_scores_corr, model_scores_rmse, predictions_by_delay = main(delais, delai_names, rain_Index)

# Affichage des résultats
for name, mae_scores, corr_scores, rmse_scores in zip(delai_names, model_scores_mae, model_scores_corr, model_scores_rmse):
    print(f"Délai {name}:")
    print(f"  MAE moyen: {np.mean(mae_scores)}")
    print(f"  Corrélation moyenne: {np.mean(corr_scores)}")
    print(f"  RMSE moyen: {np.mean(rmse_scores)}")
```
